## 不重叠时属性：fill_value，算术和reindex均可用

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)))
frame2 = frame.reindex((1, 2, 3))
frame
frame2
frame+frame2
frame.add(frame2, fill_value=0)
frame2 = frame.reindex((1, 2, 3), fill_value=100)
frame2
```

## 重命名轴索引 rename (7.2.4)
1. 可以通过函数或某种形式的映射对轴标签进行转换；
2. rename 不修改原数据（可结合字典使用）；
3. 适用于需要针对性修改时使用，若全部修改则直接进行赋值；

```python
frame = pd.DataFrame(np.arange(9).reshape(
    (3, 3)), index=list("abc"), columns=list("ABC"))
frame


def transform(x): return x.lower()
transform = lambda x: x.lower()


frame.columns.map(transform)

frame.columns = frame.columns.map(transform)
frame
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)), index=list("abc"), columns=list("ABC"))
frame
frame.rename(index=str.upper)
frame.rename(index={'a': "haha"}, columns={'C': "CCC"})
```

## 相关性和协方差
1. Series corr\cov 方法，计算两个Series中重叠的、非NA的、按索引对齐的值的相关性；
2. DataFrame 的corr和cov方法分别以DataFrame的形式返回相关性和协方差的矩阵；
3. DataFrame corrwith方法得出 DF行或列与每个列的相关性的值的Series。

```python
obj = pd.Series([4, 3, -5, 7], index=['a', 'b', 'c', 'd'])
obj3 = pd.Series([6, -5, -5, 7], index=['a', 'b', 'c', 'd'])

obj.corr(obj3)
obj.cov(obj3)
```

```python
frame = pd.DataFrame(np.random.random(12).reshape((4, 3)))
frame
frame.corr()
frame.cov()
```

```python
frame = pd.DataFrame(np.random.random(12).reshape((4, 3)))
frame.corrwith(frame.iloc[:,1], axis=1)
```

## 离散化和分箱 pd.cut pd.qcut（7.2.5）
1. 1 cats.codes 2 cats.categories  3 cats.value_counts()  4 labels
2. 等长cut precision代表精度
3. pd.qcut 分位数分箱

```python
np.random.seed(234)
age = np.random.randint(100, size=(10))
age

bin = [10, 30, 50, 70, 90, 100]

cats = pd.cut(age, bin, right=False)
cats

cats.codes

cats.categories

cats.value_counts()

group_names = ['Youth', 'YoughAdult', 'MiddleAged', 'Senior', 'Elder']
cats = pd.cut(age, bin, right=False, labels=group_names)
cats.value_counts()
```

```python
np.random.seed(234)
age = np.random.randint(100, size=(10))

pd.cut(age, 5, precision=0)
```

```python
np.random.seed(234)
age = np.random.randint(100, size=(10))

pd.qcut(age, 4) # 四分位

pd.qcut(age, [0, 0.1, 0.5, 0.8, 1]) #自定义分位数
```
