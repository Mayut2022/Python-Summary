## stack和unstack （8.1）
1. stack: 堆积、堆叠 DataFrame -》Series； unstack：拆堆 Series -》DataFrame
2. multi-index中不能包含重复数据
3. unstack() 可通过层级序号或名称来控制拆分不同层级
4. 如果是Series写为columns，则(40)->(40, 1)，则多保留一层columns名；

```python
data = pd.DataFrame(np.arange(6).reshape(2, 3), index=pd.Index(
    ['Ohio', 'Colorado'], name='state'), columns=pd.Index(['one', 'two', 'three'], name='number'))
data
```

```python
result = data.stack()
result

result.unstack(0)
result.unstack('number')
```

# 数据连接与联合 （Chapter 8.2）
> pd.merge() pd.concat() pd.combine_first()

## 合并数据，类似reindex
### pd.merge() (8.2.1) 用于有重叠列的df
1. on 参数 显示指定连接键；
2. how参数 指定连接方式，默认'inner'join, 其他有left、right、outer；
3. suffixes修改重叠列名；
4. 根据索引合并 left_index=True, right_index=True (不懂)

```python
df1 = pd.DataFrame({'key':list("aaabbcd"), 'data1':range(7)})

df2 = pd.DataFrame({'key':list("abd"), 'data2':range(3)})
df1;df2
```

```python
pd.merge(df1, df2, on='key', suffixes=('_lll', '_rrr')) # suffixes无效
pd.merge(df1, df2, left_on='data1', right_on='data2', suffixes=('_lll', '_rrr'))
```

```python
pd.merge(df1, df2, on='key', how='outer')
pd.merge(df1, df2, on='key', how='left')
pd.merge(df1, df2, on='key', how='right')
```

### DataFrame join 实例方法
> 用于按照索引合并，多个索引相同或相似但没有重叠列的DF对象，对于简单索引合并，可替代pd.concat()

```python
df2.rename(columns={'key': 'KEY'}, inplace=True)
df1.join(df2, how='inner')

df3 = df2.rename(columns={'KEY':'key3', 'data2':'data3'})
df1.join([df2, df3])
```

## 连接数据，类似rename
### pd.concat() (Chap 8.2.3)

```python
df1 = pd.DataFrame({'key':list("aaabbcd"), 'data1':range(7)})

df2 = pd.DataFrame({'key':list("abd"), 'data2':range(3)})

df3 = pd.DataFrame({'key':list("bcd"), 'data2':range(4, 7)})

df1;df2;df3
```

```python
pd.concat([df1, df3], axis=1, join='outer', keys=['one', 'two'], names=['upper', 'lower'])
pd.concat({'one':df1, 'two':df3}, axis=1, join='outer', names=['upper', 'lower'])
pd.concat([df1, df3], axis=1, join_axes=[['b', 'c']])
```

## 联合重叠数据
### pd.combine_first() 等价于 np.where()
> 既不是合并操作，也不是连接操作，两个数据集的索引全部或部分重叠。

```python
df1 = pd.DataFrame({'key': list("aaabbcd"), 'data1': range(7)})

df2 = pd.DataFrame({'key': list("abd"), 'data2': range(3)})
df1
df2
```

```python
df1.combine_first(df2)
df2.combine_first(df1)
```
