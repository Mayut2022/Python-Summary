## np.linspace np.arange 精度过高
> 在提取sif数据文件时，采用xarray提取特定经纬度数据，由于数据精度过高（3600,7200），使用np.arange和np.linspace计算机存储格式为多位小数，导致xarray无法准确定位出现报错，解决方法：使用round函数保留两位小数


```python
import numpy as np
longitude=np.linspace(-180,180,7200,endpoint=False)
latitude=np.linspace(90,-90,3600,endpoint=False)

lon=np.arange(90,-90,-0.05)

longitude.tolist()
lon.tolist()

longitude = np.array([round(i,2) for i in longitude])
latitude = np.array([round(i,2) for i in latitude])

longitude.tolist()
```

## np.random.choice 随机抽取一维数组
> [CSDN随机抽取数组](https://blog.csdn.net/a_cherry_blossoms/article/details/123439697?ops_request_misc=&request_id=&biz_id=102&utm_term=python%20%E5%A4%9A%E7%BB%B4%E6%95%B0%E7%BB%84%E9%9A%8F%E6%9C%BA%E6%8A%BD%E5%8F%96&utm_medium=distribute.pc_search_result.none-task-blog-2~all~sobaiduweb~default-2-123439697.142^v20^rank_v33,157^v15^new_3&spm=1018.2226.3001.4187)

```python
import numpy as np
import numpy.ma as ma

a = np.array([[1, 5], [3, 5], [5, 6], [7, 2], [5, 9]])

mask = a == 5
a = ma.masked_array(a, mask=mask)
a = a.filled(fill_value=-9999)

def choice():
    b = np.random.choice(a.shape[0], 5)
    c = np.random.choice(a.shape[1], 5)
    
    d = a[b, c]
    
    lat=[]
    lon=[]
    
    for i in range(len(d)):
        if d[i]==-9999:
            continue
        else:
            lat.append(b[i])
            lon.append(c[i])
            
    return lat, lon

lat, lon = choice()
```
