# 用支持插值的f-string取代C风格的格式字符串与str.format方法
## 格式化字符 dict: %(key)s表示 字符串（s）来表示dict里面名为key的那个键所保存的value.

```python
name = 'Kevin'

old = "%s loves balls. See %s shoot!" % (name, name)
print(old)
new = "%(kd)s loves balls. See %(kd)s shoot!" % {"kd": name}
print(new)

assert old == new
```

## advanced string formatting
> 不再使用%d的C风格格式说明符，有时必须转义。  help('FORMATTING')
1. 内置的format函数，逗号表示千位分隔符，^居中，>右对齐，<左对齐。
2. 许多值的格式需要调整，待调整的位置在字符串里面先用{}来替代，括号里写冒号，右边按照格式来写。
3. 可以在括号里写上数字，多个值的时候方便修改。

```python
a = format(1234.5678, ',.2f')
print(a)
b = format('MYT', '^20s')
c = format('MYT', '>20s')
print(b)
print(c)
```

```python
key = "Jason"
value = "Tatum"
formatted = '{} whynot {}'.format(key, value)
print(formatted)
formatted2 = '{:<10} whynot {:>10}'.format(key, value)
print(formatted2)
```

```python
formatted3 = '{0} loves balls. See {1} shoot!'.format(key, value)
print(formatted3)
```

## 插值格式字符串 interpolated format string, f-string  最佳选择
1. f+{},:右边表示格式，和format函数类似。
2. 可以大括号里套小括号。

```python
key = 'my car'
value = 1.234
formatted = f'{key:^10} = {value:.2f}'
print(formatted)
```

```python
places = 3
number = 1.23456
print(f'My number is {number:.1{places}f}')#13位小数
```

# 用zip函数同时遍历多个迭代器
1. zip lazy generator, 每次只生成一个元组；如果长度不一致，只要其中一个迭代完，zip就会停止。
2. 如果想按照最长的迭代器来遍历，应用内置itertools模块的 itertools.zip_longest。

```python
snacks = ['bacon', 'donut', 'muffin']
counts = [len(n) for n in snacks]
for x,y in zip(snacks, counts):
    print(f'{x} len is {y}')
```

```python
import itertools
snacks = ['bacon', 'donut', 'muffin']
counts = [len(n) for n in snacks]
snacks.append('cake')
for x, y in itertools.zip_longest(snacks, counts):
    print(f'{x} len is {y}')
```

```python
fruit = dict(apple=10, banana=8, lemon=5)
count = fruit.get('lemon', 0)
if count:  # 判断lemon的值
    print("make lemonade!")
else:
    print("out of stock!")

'''改写后'''
if count := fruit.get('lemon', 0):
    print("make lemonade!")
else:
    print("out of stock!")
    
if (count:=fruit.get('apple', 0))>=4:
    print("make apple juice!")
else:
    print("apple is out of stock!")
```
