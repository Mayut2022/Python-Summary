## np.linspace np.arange 精度过高
> 在提取sif数据文件时，采用xarray提取特定经纬度数据，由于数据精度过高（3600,7200），使用np.arange和np.linspace计算机存储格式为多位小数，导致xarray无法准确定位出现报错，解决方法：使用round函数保留两位小数


```python

longitude=np.linspace(-180,180,7200,endpoint=False)
latitude=np.linspace(90,-90,3600,endpoint=False)

lon=np.arange(90,-90,-0.05)

longitude.tolist()
lon.tolist()

longitude = np.array([round(i,2) for i in longitude])
latitude = np.array([round(i,2) for i in latitude])

longitude.tolist()
```

```python
import numpy as np
import numpy.ma as ma

a = np.array([[1, 5], [3, 5], [5, 6], [7, 2], [5, 9]])

mask = a == 5
a = ma.masked_array(a, mask=mask)
a = a.filled(fill_value=-9999)

def choice():
    b = np.random.choice(a.shape[0], 5)
    c = np.random.choice(a.shape[1], 5)
    
    d = a[b, c]
    
    lat=[]
    lon=[]
    
    for i in range(len(d)):
        if d[i]==-9999:
            continue
        else:
            lat.append(b[i])
            lon.append(c[i])
            
    return lat, lon

lat, lon = choice()
```

## 数组类型转换

```python
import numpy as np

# 创建一个示例数组
original_array = np.array([1.5, 2.7, 3.3, 4.8])

# 将数组转换为整数类型 (int)
integer_array = original_array.astype(np.int32)

# 将数组转换为无符号整数类型 (uint)
uint_array = original_array.astype(np.uint8)

# 将数组转换为浮点数类型 (float)
float_array = original_array.astype(np.float64)

# 输出不同数据类型的数组
print("Original Array:", original_array)
print("Integer Array:", integer_array)
print("Unsigned Integer Array:", uint_array)
print("Float Array:", float_array)
```

## np.repeat() 
> numpy.repeat(a, repeats, axis=None)

```python
import numpy as np

# 示例 1：重复一维数组中的元素
arr = np.array([1, 2, 3])
repeated = np.repeat(arr, 3)
print(repeated)
# 输出：[1 1 1 2 2 2 3 3 3]

# 示例 3：按轴重复
matrix = np.array([[1, 2], [3, 4]])
repeated = np.repeat(matrix, 3, axis=0)  # 沿行的方向重复
print(repeated)

# 示例 4：使用数组作为 repeats 参数
arr = np.array([1, 2, 3])
repeats = np.array([2, 3, 1])
repeated = np.repeat(arr, repeats)
print(repeated)
# 输出：[1 1 2 2 2 3]
```

## Difference between a=b & a=b.copy()
### data = data2 与 data = data2.copy() 的区别是？
> **data=data2**  
> 这是一种浅复制，修改其中一个变量的内容，另一个也会反映更改
> 该操作将data2变量简单地指向了data2所引用的对象，二者实际上引用相同的内存空间，因此它们是共享数据的  

> **data=data2.copy()**  
> 这是一种深复制（deep copy）操作。它会创建一个完全独立于原始对象的新对象，新对象的内容与原始对象相同;修改其中之 一不会影响另一个变量的内容

```python
data2 = np.array([1, 2, 3])
data = data2  # 这只是创建了一个引用，没有复制数据
data[0] = 100
print(data2)  # 输出 [100, 2, 3]，因为两者引用相同的数据

result = (data is data2)
result

data2 = np.array([1, 2, 3])
data = data2.copy()  # 创建了一个新的独立复制
data[0] = 100
print(data2)  # 输出 [1, 2, 3]，因为它们是独立的对象

result = (data is data2)
result
```

### is运算符：判断两数组是否来自相同的数据空间

```python
a = [1, 2, 3]
b = a  # b引用了a的对象

# 使用is运算符检查是否引用相同的对象
result = (a is b)
print(result)  # 输出 True，a和b引用同一个对象

c = [1, 2, 3]
# 使用is运算符检查c是否引用相同的对象
result = (c is a)
print(result)  # 输出 False，c和a引用不同的对象
```
