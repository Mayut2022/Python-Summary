```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# dodge参数
> [dodge参数](http://seaborn.pydata.org/generated/seaborn.barplot.html)  
> When hue mapping is used, whether elements should be narrowed and shifted along the orient axis to eliminate overlap. If "auto", set to True when the orient variable is crossed with the categorical variable or False otherwise.

```python
data = sns.load_dataset("tips")

# 使用 dodge 参数将 day 列的不同取值分组展示
sns.barplot(x="day", y="total_bill", data=data, dodge=False)

plt.show()
```

# order参数改变x坐标轴显示的顺序

```python
# 创建一个示例DataFrame
data = {'LagMonths': [2, 1, 3, 4],
        'Slope': [0.5, 0.8, 0.4, 0.9],
        'Year': [2019, 2020, 2019, 2020]}
df = pd.DataFrame(data)

# 指定x坐标轴顺序
x_order = [1, 2, 3, 4]

# 绘制条形图
fig, ax = plt.subplots()
sns.barplot(ax=ax, data=df, x="LagMonths", y="Slope", hue="Year", palette=["b", "r", 'lightgray'], order=x_order)

# 显示图形
plt.show()
```
