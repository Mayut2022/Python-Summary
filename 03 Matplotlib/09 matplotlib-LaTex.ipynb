# 常用设置rcParams及LeTax的格式
除了斜体（italic）和加粗（bold）之外，还有一些其他常用的设置可以应用于 matplotlib 中的文本和数学表达式。

在 matplotlib 中，可以使用 rcParams 对象来设置全局参数，包括文本和数学表达式的格式。rcParams 是一个字典样式的对象，用于存储和管理 matplotlib 的所有配置选项。

以下是一些常用的参数和对应的格式：

font.family：设置字体族（font family），例如 'serif'、'sans-serif'、'monospace' 等。
font.style：设置字体样式（font style），例如 'normal'（正常）、'italic'（斜体）等。
font.variant：设置字体变体（font variant），例如 'normal'（正常）、'small-caps'（小型大写字母）等。
font.weight：设置字体粗细（font weight），例如 'normal'（正常）、'bold'（加粗）等。
font.size：设置字体大小（font size），例如 10、12 等。
mathtext.fontset：设置数学文本的字体集（mathematical text font set），例如 'dejavusans'、'dejavuserif'、'cm'、'stix' 等。
mathtext.rm：设置数学文本的正体字体（roman font），例如 'serif'、'sans-serif' 等。
mathtext.it：设置数学文本的斜体字体（italic font），例如 'serif:italic'、'sans-serif:italic' 等。
mathtext.bf：设置数学文本的加粗字体（bold font），例如 'serif:bold'、'sans-serif:bold' 等。
这里提到的格式是一种简化的描述方式，具体的格式取决于所使用的字体和 LaTeX 引擎。不同的字体和引擎可能具有不同的格式设置方式。

需要注意的是，对于 LaTeX 的格式，可以使用类似 '\textit{...}'、'\textbf{...}'、'\mathrm{...}' 等命令来指定斜体、加粗和正体等。具体的格式设置方法可以参考 LaTeX 的相关文档和语法。

通过修改 rcParams 中的相应参数，您可以自定义 matplotlib 图形中文本和数学表达式的格式。

1. 在 LaTeX 表达式中使用 $...$ 来表示数学模式
2. 数学表达式和正常文本字体的管理是区分开的

在上述代码中，config 是一个字典，其中包含了一些用于设置 matplotlib 图形的参数。

以下是这些参数的解释：

"font.family": 'serif'：这个参数设置了字体族（font family），指定了在图形中使用的字体类型。这里将其设置为 'serif'，表示使用衬线字体（serif font），如 Times New Roman 或 Computer Modern Serif。

"font.size": 10.5：这个参数设置了字体大小（font size），指定了在图形中使用的字体的基本大小。这里将其设置为 10.5。

"mathtext.fontset": 'stix'：这个参数设置了数学文本的字体集（mathematical text font set），指定了在数学公式中使用的字体集。这里将其设置为 'stix'，表示使用 STIX 字体集，它包含了一系列数学符号和字母。

"font.serif": ['times new roman']：这个参数设置了衬线字体（serif font），指定了在图形中使用的衬线字体的名称。这里将其设置为 ['times new roman']，表示使用 Times New Roman 字体作为衬线字体。

rcParams.update(config) 将这些参数应用于 matplotlib 的全局配置（rc）。

在 matplotlib 中，"mathtext.fontset" 参数用于设置数学文本的字体集。该参数控制在数学环境中使用的字体类型。

该参数的默认值是 "dejavusans"，表示使用 DejaVu Sans 字体集。DejaVu 字体是一组开源字体，包含了广泛的 Unicode 字符覆盖范围，适用于显示数学符号和字母。

除了 "dejavusans" 外，"mathtext.fontset" 还支持以下可选参数：

"dejavusans"：使用 DejaVu Sans 字体集。
"dejavuserif"：使用 DejaVu Serif 字体集。
"cm"：使用 Computer Modern 字体集。Computer Modern 是 LaTeX 默认的字体，常用于科学出版物。
"stix"：使用 STIX 字体集。STIX 字体是专门为数学排版设计的字体，包含了丰富的数学符号和字母。
"stixsans"：使用 STIX Sans 字体集。STIX Sans 是 STIX 字体的无衬线版本。
"custom"：使用自定义字体集。您可以通过设置 "mathtext.fontset" 参数为 "custom"，然后在 "mathtext.fontset_custom" 参数中指定自定义字体的名称。

```python
% config InlineBackend.figure_format = 'retina'
% matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
 
from matplotlib import rcParams
 
config = {
            "font.family": 'serif',
            "font.size": 10.5,
            "mathtext.fontset": 'stix',
            "font.serif": ['SimSun'],
         }
rcParams.update(config)
 
x=[1,2,3]
y=[1,2,3]

plt.figure(1, dpi=300)
plt.plot(x, y,label="时间随着距离的变换参数：y=0.5*x")
plt.plot(x, y, label=r'$\mathit{label}_{\mathrm{sub}}$')
plt.xlabel('Time (s)')
plt.ylabel('Force ($\mathregular{F_{net}^{max}}$)')
plt.legend() # 显示图例
plt.show()
```

> 如何将数学公式中的字体自定义为 times new roman？

```python
import matplotlib.pyplot as plt

plt.rcParams["mathtext.fontset"] = "custom"
plt.rcParams["mathtext.rm"] = "Times New Roman"
plt.rcParams["mathtext.it"] = "Times New Roman:italic"
plt.rcParams["mathtext.bf"] = "Times New Roman:bold"

plt.title(r'This is a math expression: $x^2 + y^2 = z^2$', fontsize=14)

plt.show()
```

1. rm: roman font 正体字体
2. it: italic font 斜体字体
3. bf: bold font 加粗字体

```python
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt

plt.rcParams["mathtext.fontset"] = "custom"
plt.rcParams["mathtext.rm"] = "Times New Roman"
plt.rcParams["mathtext.it"] = "Times New Roman:italic"
plt.rcParams["mathtext.bf"] = "Times New Roman:bold"

# 创建示例图形
fig, ax = plt.subplots(dpi=300)

# 设置斜体文本
text = r'$\mathit{italic\ text}$ and normal text'
# text = 'normal text'

# 在图形中显示斜体文本
ax.text(0.5, 0.5, text, fontsize=14)
plt.title(r'F$\it{abc}$',fontsize=25,fontweight='bold',fontfamily='Times New Roman')

# 显示图形
plt.show()
```
