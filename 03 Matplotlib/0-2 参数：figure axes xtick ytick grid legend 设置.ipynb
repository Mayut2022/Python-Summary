# 修改tick_params()
> length, width 代表刻度的长短、宽窄；pad代表字和刻度的间隔

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

# 创建一个示例图形
fig, ax = plt.subplots()
ax.tick_params(labelsize=12, length=10, pad=2, labelcolor='red')
ax.set_title("Area Percentage", fontsize=15, loc="left", pad=10)

# 添加逐年统计子图
sub_ax = fig.add_axes([0.2, 0.35, 0.5, 0.35])
year = np.arange(0, 9)
data2 = np.linspace(0, 0.6, 9)

sub_ax.bar(year, data2, color='#FF0000', width=0.8)
sub_ax.tick_params(labelsize=10)
sub_ax.set_xlim(-0.5, 8.5)
sub_ax.set_xticks(np.arange(0, 9, 2))
sub_ax.set_xticklabels(np.arange(0, 9, 2), color="red", fontweight="bold")
sub_ax.set_ylim(0, 0.6)
sub_ax.set_yticks([0, 0.25, 0.5])
sub_ax.set_yticklabels(["0", "25%", "50%"], fontweight="bold")
sub_ax.set_title("Area Percentage", fontsize=10, loc="left")
plt.show()
```
