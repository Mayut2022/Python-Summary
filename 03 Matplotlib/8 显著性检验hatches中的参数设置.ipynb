## 简单修改，使用全局变量进行修改

```python
plt.rcParams['hatch.color']='orange'

import matplotlib.pyplot as plt
import numpy as np
# 随机生成一个二维数据用于绘制图形
x = np.linspace(-3, 5, 150).reshape(1, -1)
y = np.linspace(-3, 5, 120).reshape(-1, 1)
z = np.cos(x) + np.sin(y)
x, y = x.flatten(), y.flatten()
#x,y,z的形状分别为(150,) (120,) (120, 150)


plt.figure(dpi=200)
n_levels = 6  # 6个水平层级，对应8个填充区间
plt.contour(x, y, z, n_levels, colors='black', linestyles='-')
cs = plt.contourf(
    x, y, z, n_levels, colors='none',
    hatches=['.', '/', '\\', None, '\\\\', '*'],
    extend='lower',)
```

## 细致修改
> 在Matplotlib中，collection是指等高线图（contour plot）或填充等高线图（contourf plot）中的图形集合对象。它表示一组具有相同属性（例如颜色、填充样式等）的图形元素，可以是线条、多边形或路径等。 

> 在你的代码中，cs.collections是一个包含所有填充区间的集合对象列表。每个集合对象表示一个填充区间，包含该区间内的所有图形元素。你可以通过遍历这个列表来访问每个填充区间的集合对象。  

> 通过访问集合对象，你可以对其应用各种属性设置，例如设置边缘颜色、填充颜色、线宽等。在你的代码中，通过循环for collection in cs.collections，你可以逐个访问每个填充区间的集合对象，并对其进行属性设置，如collection.set_edgecolor(colors[i % len(colors)])和collection.set_linewidth(0)。  

> 总而言之，集合对象（collection）是Matplotlib中用于表示图形元素集合的对象，它使得对这些元素进行统一设置和操作变得更加方便。

```python
import matplotlib.pyplot as plt
import numpy as np
# 随机生成一个二维数据用于绘制图形
x = np.linspace(-3, 5, 150).reshape(1, -1)
y = np.linspace(-3, 5, 120).reshape(-1, 1)
z = np.cos(x) + np.sin(y)
x, y = x.flatten(), y.flatten()
#x,y,z的形状分别为(150,) (120,) (120, 150)


plt.figure(dpi=200)
n_levels = 6  # 6个水平层级，对应8个填充区间
plt.contour(x, y, z, n_levels, colors='black', linestyles='-')
cs = plt.contourf(
    x, y, z, n_levels, colors='none',
    hatches=['.', '/', '\\', None, '\\\\', '*'],
    extend='lower',)
# 8个填充区间的填充颜色
colors = ['maroon', 'red', 'darkorange', 'gold', 'forestgreen',
          'darkturquoise', 'dodgerblue', 'darkviolet']

# 指定每个区间形状填充的颜色
for i, collection in enumerate(cs.collections):
    collection.set_edgecolor(colors[i % len(colors)])

# # 将边界填充粗细设置为0
for collection in cs.collections:
    collection.set_linewidth(0)
```
