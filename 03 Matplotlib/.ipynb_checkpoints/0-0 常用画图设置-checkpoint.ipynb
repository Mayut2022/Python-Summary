# 上下标设置
1. $()$, 上标：^，下标：_ , 如果上下标多余一个字符，则必须用花括号括起来。
2. 加上\mathregular，才能确保上下标同一个字体。（数字字母通用）

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['times new roman']  # 指定默认字体

x1 = np.linspace(0, 10, 50)
x2 = np.linspace(0, 15, 100)
y1 =np.sin(x1)
y2 =np.cos(x2)

plt.figure(figsize=(2, 2), dpi=300)
plt.plot(x1,y1,label='sin$_{\mathregular{xxyyzz}}$')
plt.plot(x2,y2,label='cos$^x$')
plt.xlabel('2020$^{2021}$')
plt.ylabel('N uptake rate (kg N ha$^{\mathregular{-1}}$yr$^{\mathregular{-1}}$)',fontproperties = 'Times New Roman')
plt.legend()
```

# 多子图中设置统一的ylabel

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(4, 4, figsize=(8, 6), dpi=300)

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)

fig.text(0., 0.5, 'Y Label', va='center_baseline', rotation='vertical')

plt.tight_layout()
plt.show()
```
