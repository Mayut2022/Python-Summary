## DataFrame 柱状图
### 干旱严重程度统计
1. index为横坐标，columns为每个横坐标的子bar，value值为纵坐标；

```python
# %%
df = pd.read_excel(
    r"E:\SPEI_base\python NorthEast Asia region1\region1 干旱统计.xlsx", sheet_name="Drought all")

ds = df.iloc[:, 6:]
ds_num = ds.iloc[:, [0, 2, 4, 6]]
ds_per = ds.iloc[:, [1, 3, 5, 7]]

# %%


def plot(data, title):
    fig = plt.figure(figsize=(18, 6), dpi=500)
    fig.subplots_adjust(left=0.05, bottom=0.15, right=0.95,
                        top=0.92, wspace=None, hspace=0.1)
    ax = fig.add_subplot(111)

    color = ["yellow", "orange", "red", "gray"]

    data.plot.bar(rot=0, ax=ax, color=color, width=0.6, alpha=0.8)

    ax.set_ylim(0, 0.8)
    ax.set_yticks(np.arange(0, 0.81, 0.1))
    #ax.set_ylabel("percent", fontsize=15)
    ax.tick_params(labelsize=10)

    plt.suptitle(f'{title}', fontsize=15)
    #plt.savefig(rf'E:/SPEI_base/JPG_NorthEast Asia region1/干旱百分比/{title}.jpg', bbox_inches='tight')
    plt.show()


# %%
ds_per.columns = ['Mild', 'Moderate', 'Severe', 'Sum area']
plot(ds_per, "Drought Severity Percentage")
# %%
ds_per
```

```python
def plot2(data, title):
    fig = plt.figure(figsize=(18, 6), dpi=500)
    fig.subplots_adjust(left=0.05, bottom=0.15, right=0.95,
                        top=0.92, wspace=None, hspace=0.1)
    ax = fig.add_subplot(111)

    color = "dodgerblue"

    data.plot.bar(rot=0, ax=ax, color=color, width=0.35, alpha=0.8)

    ax.set_ylim(0, 8)
    ax.set_yticks(np.arange(0, 8.1, 1))
    #ax.set_ylabel("percent", fontsize=15)
    ax.tick_params(labelsize=10)

    plt.suptitle(f'{title}', fontsize=15)
    #plt.savefig(
     #   rf'E:/SPEI_base/JPG_NorthEast Asia region1/干旱百分比/{title}.jpg', bbox_inches='tight')
    plt.show()


# %%
dd = df['DD']
plot2(dd, "Duration Month")
```

### 干旱程度和干旱时间多子图绘制

```python
def plot3(data1, data2, title):
    fig = plt.figure(figsize=(18, 13), dpi=500)
    fig.subplots_adjust(left=0.05, bottom=0.1, right=0.95,
                        top=0.95, wspace=None, hspace=0.06)
    axs = fig.subplots(2, 1, sharex=True)

    color1 = "dodgerblue"
    data1.plot.bar(rot=0, ax=axs[0], color=color1, width=0.35, alpha=0.8)

    color2 = ["yellow", "orange", "red", "gray"]
    data2.plot.bar(rot=0, ax=axs[1], color=color2, width=0.6, alpha=0.8)

    axs[0].set_ylim(0, 8)
    axs[0].set_yticks(np.arange(0, 8.1, 1))
    axs[1].set_ylim(0, 0.8)
    axs[1].set_yticks(np.arange(0, 0.81, 0.1))
    axs[1].legend(fontsize=15)

    for ax in axs:
        ax.tick_params(labelsize=15)

    plt.suptitle(f'{title}', fontsize=20)
#     plt.savefig(
#         rf'E:/SPEI_base/JPG_NorthEast Asia region1/干旱百分比/{title}.jpg', bbox_inches='tight')
    plt.show()


# %%
plot3(dd, ds_per, "DM and DS")
```
